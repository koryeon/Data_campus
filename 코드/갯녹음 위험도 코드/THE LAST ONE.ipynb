```python
metrics = []

for i in range(0, 1): 
#for i in range(0, 161): 
    dp = pd.DataFrame(main2[i])
    dp['ds'] = list(pd.date_range('1900-01-01', periods=76))
    for j in range(len(col2)):
        dp2 = dp[['ds', col2[j]]]
        dp2.columns = ['ds', 'y']

        robustScaler = RobustScaler()
        temp = np.array(dp2['y']).reshape((len(dp2['y']), 1))
        train_data_robustScaled = robustScaler.fit_transform(temp)
        train_data_robustScaled_df = pd.DataFrame(train_data_robustScaled)
        train_data_robustScaled_df['ds']= pd.to_datetime(dp2['ds'])
        train_data_robustScaled_df.columns = ['y', 'ds']
        
        model = Prophet(growth='linear',
                        changepoint_prior_scale=0.5,
                        seasonality_mode= 'multiplicative',
                        daily_seasonality=False,
                        weekly_seasonality=False,
                        yearly_seasonality=False,
                        ).add_seasonality(
                        name='quarterly',
                        period=4,
                        fourier_order=55,#25
                        prior_scale=20)
        model.fit(train_data_robustScaled_df)
        
        # train set 마지막 1년 날짜 생성
        last_1year = list(pd.date_range('1900-01-01', periods=80))
        last_1year = pd.DataFrame(last_1year, columns = ['ds'])
        last_1year['ds']= pd.to_datetime(last_1year['ds'])
        
        forecast = model.predict(last_1year)
        
        future = []
        future.append(station_list[i])
        future.append(col2[j])
        rmse = np.sqrt(mean_squared_error(train_data_robustScaled_df['y'], forecast['yhat'][:76]))
        r2 = r2_score(train_data_robustScaled_df['y'], forecast['yhat'][:76])
        future.append(rmse)
        future.append(r2)
        
        temp2 = np.array(forecast['yhat'][76:]).reshape((len(forecast['yhat'][76:]), 1))
        t2_prime_prime = robustScaler.inverse_transform(temp2)
        
        future.append(t2_prime_prime.mean())
    
        print(station_list[i], "future:", future)

        metrics.append(future)
```
